Importing the data and reading all the sheets using a for loop

```python
import pandas as pd
filepath='/Users/isabellareeser/similar_storms_nonstorms.xlsx'
f=pd.ExcelFile(filepath)
list_of_dfs=[]

for sheet in f.sheet_names:
    df=f.parse(sheet)
    list_of_dfs.append(df)
    
data=list_of_dfs
```

Splitting up the data into storm and no storm for the graphs.

```python
storm=[]
nostorm=[]
for i in range(len(data)):
    if (i%2)==0:
        storm.append(data[i])
    else:
        nostorm.append(data[i])
```

```python
import matplotlib.pyplot as plt
```
